### # of upper and lower charactes 

```python
Lista = ["LLLs", "Martes", "Miercoles","jueves"]
def LOUP (Lista):
    print("insert index")
    j = int(input())
    a = int ()
    b = int ()
    for c in range(len (Lista[j])):
        x = Lista [j][c].islower ()
        if x == True:
            a = a + 1
        elif x == False:
            b = b + 1
    print ("# of upper" , b)
    print ("# of lower" , a)
     
LOUP(Lista)  
```
